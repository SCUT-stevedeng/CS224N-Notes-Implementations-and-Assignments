## Backpropagation
Backpropagation is based on forward propagation, during which neural models generate outputs. These outputs are then fed into a loss function to measure the modelâ€™s performance. To optimise the model, gradient descent is commonly used: we compute the gradients, which represent the direction of the steepest ascent of the loss function, and move in the opposite direction to minimise it. Backpropagation updates the model parameters towards this descending direction, usually scaled by a hyperparameter called the **learning rate**.
