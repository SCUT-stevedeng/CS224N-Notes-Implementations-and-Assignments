## Backpropagation
Backpropagation is based on forward propagation, during which neural models generate outputs. These outputs are then fed into a loss function to measure the model’s performance. To optimise the model, gradient descent is commonly used: we compute the gradients, which represent the direction of the steepest ascent of the loss function, and move in the opposite direction to minimise it. Backpropagation updates the model parameters towards this descending direction, usually scaled by a hyperparameter called the **learning rate**.

$$θ = θ - η · ∇θ J(θ)$$

```python
import torch
import torch.nn as nn
```

```python
layer1=torch.randn((20,8), requires_grad=True)
layer2=torch.randn((8,1), requires_grad=True)

x=torch.randn((10, 20)) #shape:(batch_size, input_features)
y=torch.randint(0,2,(10,1))

output=x@layer1@layer2

prob=torch.sigmoid(output)
eplison=1e-8
lr=1e-2
loss=(-(y*torch.log(prob+eplison)+(1-y)*torch.log(1-prob+eplison))).mean()
```

```python
loss.backward()
with torch.no_grad():
    layer1-= lr * layer1.grad
    layer2-= lr * layer2.grad
```
